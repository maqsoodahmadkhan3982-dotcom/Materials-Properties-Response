```python
# Problem: Solve the system 5x+3y=3;    2x-y=4
# The Augmented Matrix of the system is
import numpy as np
Ag=np.array([[5, 3, 3],
            [2,-1,4]])
# R1-2R2
print(Ag)

Ag1=np.array([Ag[0]-2*Ag[1],Ag[1]])
print(Ag1)
```

```python
# R2-2R1
Ag2=np.array([Ag1[0],Ag1[1]-2*Ag1[0]])
print(Ag2)
```

```python


# -1/11*R2
Ag2=np.array([Ag1[0],Ag1[1]-2*Ag1[0]])
print(Ag2)

Ag3=np.array([Ag2[0],-1/11*Ag2[1]])
print(Ag3,'\n')

# R1-R2
Ag4=np.array([Ag3[0]-5*Ag3[1],Ag3[1]])
print(Ag4)

print('The Solution is given by',f'\nx={Ag4[0][2]},y={Ag4[1][2]}')
```

```python
# Using function rref from sympy
import sympy as sp
Ag=sp.Matrix([[5, 3, 3],
            [2,-1,4]])
Ag, pivot_columns=Ag.rref()
print(Ag)
print('The Solution is given by',f'\nx={Ag.row(0)[2]},y={Ag.row(1)[2]}')

print(f'The Solution is given by \n x={Ag.row(0)[2]},y={Ag.row(1)[2]}')

print(f'The pivot columns are {pivot_columns}')

print(f'The Rank of Ag={len(pivot_columns)}')
```

```python
# Using function rref from sympy
import sympy as sp
Ag=sp.Matrix([[5, 3, 3],
            [2,-1,4]])
Ag, pivot_columns=Ag.rref()
Ag=np.array(Ag)                      #Converting a unmutable object that is tuple into a List, a mutable object
Ag=Ag.astype(float)                 #Changing type of data so that we can get results in decimal form
print(Ag)
print('The Solution is given by',f'\nx={Ag[0][2]},y={Ag[1][2]}')
```

# System of Three Linear Equations Linear System:

#  x-y+2z=5

# 2x-2y+4z=10

# 3x-3y+6z=15

```python
A=np.array([[1,1,2],[2,8,10],[3,7,6]])
B=np.array([5,0,15])
X=np.linalg.solve(A,B)
print('The solution of the system is (x,y,z)=',X)
```

```python
Ag=np.array([[1,-1,2,5],[2,-2,4,10],[3,-3,6,15]])

Ag=np.array([Ag[0],Ag[1]-2*Ag[0],Ag[2]-3*Ag[0]])
print(Ag)
```

```python
import sympy as sp
Ag=sp.Matrix([[1,-1,2,5],[2,-2,4,10],[3,-3,6,15]])
Ag, pivot_columns=Ag.rref()
print(Ag)
print(f'The pivot columns are {pivot_columns}')

print(f'The Rank of Ag={len(pivot_columns)}')

print(f'Total number of free variables are {len(Ag.row(0))-1-len(pivot_columns)}')
```

```python
import sympy as sp
import numpy as np
Ag=sp.Matrix([[1,1,2,9],[2,4,-3,1],[3,6,-5,0]])
RREF,Pivot=Ag.rref()
AG=np.array(RREF)
print('The Row Reduced Echelon Form of Augmented Matrix is:',f'\n{RREF}')
print('The Augmented matrix as an array is ',f'\n{AG}')
print('The Solution is given by',f'\nx={AG[0][3]},y={AG[1][3]},z={AG[2][3]}')
print('The Pivot columns are: \n',np.array(Pivot))
```

```python
import sympy as sp


I = sp.Matrix(I)                                         # Define a matrix with floating-point entries


I_as_fractions = I.applyfunc(sp.nsimplify)             # Convert floating-point entries to fractions

I1=np.array(I_as_fractions)

print(I_as_fractions ,'\n')
print(I1,'\n')
```

```python
# Using Linalgebra Package
P=np.linalg.multi_dot([np.array([[1,1,1],[4,2,1],[1,-1,1]]),I1])
print(P)
#print(np.abs(np.round(P)))
```
