# **Question 1**

4️⃣ Newton Method (Numerical Derivative Version)

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

def f(x):
    return 50*x**2-100000

def df(x):
    h=1e-6
    return (f(x+h)-f(x-h))/(2*h)

X=[]
ERR=[]

x0=20
n=10
Es=0.5*(10**(2-n))*100

print("The required error should stay less than",np.round(Es,10))

X.append(x0)

for i in range(50):

    xn=x0-f(x0)/df(x0)
    X.append(xn)

    ERROR=np.abs((xn-x0)/xn)*100
    ERR.append(ERROR)

    if ERROR<Es:
        print("The required accuracy is achieved at",i,"step because error is",ERROR,"& Es is",Es)
        break
    else:
        x0=xn

print('The solution (x) and value of f(x) are:',[X[len(X)-1],f(X[len(X)-1])])
print("The approximate solutions at all steps are:",X)

plt.plot(range(1,len(ERR)+1),ERR,marker='o')
plt.xlabel("Iterations")
plt.ylabel("Percent Error")
plt.title("Newton Method Error vs Iterations")
plt.grid()
plt.show()
```

2️⃣ Bisection Method (Same Output Style)

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

x=sp.symbols('x')
y=50*x**2-100000

f=sp.lambdify(x,y)

a=0
b=100

X=[]
ERR=[]

n=10
Es=0.5*(10**(2-n))*100

print("The required error should stay less than",np.round(Es,10))

for i in range(50):

    c=(a+b)/2
    X.append(c)

    if i>0:
        ERROR=np.abs((X[i]-X[i-1])/X[i])*100
        ERR.append(ERROR)

        if ERROR<Es:
            print("The required accuracy is achieved at",i,"step because error is",ERROR,"& Es is",Es)
            break

    if f(a)*f(c)<0:
        b=c
    else:
        a=c

print('The solution (x) and value of f(x) are:',[X[len(X)-1],f(X[len(X)-1])])
print("The approximate solutions at all steps are:",X)

plt.plot(range(1,len(ERR)+1),ERR,marker='o')
plt.xlabel("Iterations")
plt.ylabel("Percent Error")
plt.title("Bisection Error vs Iterations")
plt.grid()
plt.show()
```

1️⃣ Bisection Method

```python
import numpy as np
import matplotlib.pyplot as plt

# constants
i0 = 1e-6
i = 1e-4
beta = 0.12
R = 8.314
T = 298

def f(E):
    return i0*np.exp((beta*E)/(R*T)) - i

# stopping criteria
Es = 5e-7
max_iter = 100

xl = 90000
xu = 100000

X=[]
ERR=[]

for k in range(max_iter):

    xr = (xl+xu)/2
    X.append(xr)

    if k>0:
        error = abs((X[k]-X[k-1])/X[k])*100
        ERR.append(error)

        if error < Es:
            print("The required accuracy is achieved at",k,"step because error is",error,"& Es is",Es)
            break

    if f(xl)*f(xr) < 0:
        xu = xr
    else:
        xl = xr

print("The solution (E) and value of f(E) are:",[xr,f(xr)])
print("The approximate solutions at all steps are:",X)

plt.plot(range(1,len(ERR)+1),ERR,marker='o')
plt.xlabel("Iterations")
plt.ylabel("Percent Error")
plt.title("Bisection Method Error vs Iterations")
plt.grid()
plt.show()
```
